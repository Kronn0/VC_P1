TAREA: Modifica de alguna forma los valores de un plano de la imagen

```python
vid = cv2.VideoCapture(0)

#Convertimos BGR a HSV
blue = np.uint8([[[255,0,0]]])
green = np.uint8([[[0,255,0]]])
red = np.uint8([[[0,0,255]]])
hsv_blue = cv2.cvtColor(blue,cv2.COLOR_BGR2HSV)
hsv_green = cv2.cvtColor(green,cv2.COLOR_BGR2HSV)
hsv_red = cv2.cvtColor(red,cv2.COLOR_BGR2HSV)

#Definimos el rango de los colores en HSV
lower_blue = np.array([110,100,20])
upper_blue = np.array([130,255,255])
lower_green = np.array([50,100,20])
upper_green = np.array([70,255,255])
lower_red = np.array([0,100,20])
upper_red = np.array([10,255,255])

while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    # Convertimos BGR a HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Creamos una máscara para cada color
    blue_mask = cv2.inRange(hsv, lower_blue, upper_blue)
    green_mask = cv2.inRange(hsv, lower_green, upper_green)
    red_mask = cv2.inRange(hsv, lower_red, upper_red)

    # Hacemos un AND a cada bit de la imagen para cada color
    blue_res = cv2.bitwise_and(frame,frame, mask=blue_mask)
    green_res = cv2.bitwise_and(frame,frame, mask=green_mask)
    red_res = cv2.bitwise_and(frame,frame, mask=red_mask)

    h, w, _ = frame.shape

    # Juntamos las tres variantes para una comparación más fácil
    res_collage = np.hstack((blue_res, green_res, red_res))
    mask_collage = np.hstack((blue_mask, green_mask, red_mask))
    
    # Muestra fotograma redimensionando a un tercio para comparar las 3 vistas
    cv2.imshow('frame', cv2.resize(frame, (int(w/3),int(h/3)),cv2.INTER_NEAREST))
    cv2.imshow('mask', cv2.resize(mask_collage, (int(w*1.5),int(h/2)),cv2.INTER_NEAREST))
    cv2.imshow('res', cv2.resize(res_collage, (int(w*1.5),int(h/2)),cv2.INTER_NEAREST))

    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Evento de ratón para mostrar el valor RGB del píxel

```python
# Función que trata eventos de ratón 
def mouse_events(event, x, y, flags, params):
    global px, py, R, G, B
    px = -1
    # Movimiento del puntero     
    if event==cv2.EVENT_MOUSEMOVE:
        #Valores del píxel
        B = frame[y, x, 0]
        G = frame[y, x, 1]
        R = frame[y, x, 2]  

        px = x
        py = y

     
  
```

```python
vid = cv2.VideoCapture(0)

# Tipografía para mostrar texto
font = cv2.FONT_HERSHEY_SIMPLEX
  
lanzado = 0
px = -1
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret: 
        # Activa em manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)   
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1
            
        # Muestra valores RGB intentando centrar en el puntero
        if px > -1:
            cv2.putText(frame, '{}'.format(R), (px-45,py-5), font, 0.5, (0, 0, 255), 1)
            cv2.putText(frame, '     {}'.format(G), (px-54,py-5), font, 0.5, (0, 255, 0), 1)
            cv2.putText(frame, '         {}'.format(B), (px-54,py-5), font, 0.5, (255, 0, 0), 1)
            
        cv2.imshow('Cam', frame)   
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Pintar el circulo en las zonas 8x8 más clara y oscura

```python
vid = cv2.VideoCapture(0)
  

RGB = 0
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    
    if ret: 
         #pasamos la imagen a una imagen en gris. 
        frame_gris = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #Extraemos el tamaño de la imagen
        height, width = frame_gris.shape
        #Fijamos el tamaño de la zona que estamos buscando
        region_size = 8
        #definimos una intensidad muy alta, en este caso ininito porque será inalcanzable
        min_intensity = int('inf')
        #definimos una intensidad muy baja, en este caso -1 porque será inalcanzable
        max_intensity = -1
        #definimos los valores del primer elemento de la zona oscura y clara
        min_x, min_y = 0,0
        max_x, max_y = 0,0
        #recorremos la matriz desde 0 hasta el fin de la matriz menos el tamaño de esta (más uno para evitar out of bounds)         
        for x in range(0, height - region_size + 1, region_size):
            for y in range(0, width - region_size + 1, region_size):
                #hacemos una matriz de 8x8 de la region que vamos a analizazr
                region = frame_gris[x:x+region_size, y:y+region_size]
                #medimos la intensidad de la region con la media de esta
                intensity = np.mean(region)
                #si la intensidad es menor que la minima actual se actualiza
                if intensity < min_intensity:
                    min_intensity = intensity
                    min_x, min_y = x, y
                #si la intensidad es mayor que la maxima actual se actualiza
                if intensity > max_intensity:
                    max_intensity = intensity
                    max_x,max_y = x,y

        
        #pintamos un circulo negro en la zona más oscura
        cv2.circle(frame, (min_y, min_x),6, (0, 0, 0), -1)  
        #pìntamos un circulo blanco e la zona más clara 
        cv2.circle(frame, (max_y, max_x),6, (255, 255, 255), -1)  
        
        
       #cv2.circle(frame,min_loc,4,(255,0,0),-1); 
        #cv2.circle(frame,max_loc,4,(0,255,0),-1); 
        
        cv2.imshow('Cam', frame) 
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Pop art
Salida de la webcam "inspirada" en Andy Warhol

```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cámara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resolución para ocupar menos pantalla
w=int(w/2)
h=int(h/2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w) #En Mac no reacciona a estos comandos
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*2,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
bl = collage[h:h+h,0:w]
br = collage[h:h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frameIN = vid.read()

    #Menor tamaño
    frame = cv2.resize(frameIN, (int(w),int(h)),cv2.INTER_NEAREST)

    if ret:
        #Separamos canales
        r = frame[:,:,2]
        g = frame[:,:,1]
        b = frame[:,:,0]

        #Jugamos
        tl[:,:,0] = b
        tl[:,:,1] = g
        tl[:,:,2] = r

        tr[:,:,0] = 255 - r
        tr[:,:,1] = g
        tr[:,:,2] = b
        
        bl[:,:,0] = r
        bl[:,:,1] = 255 - b
        bl[:,:,2] = g

        br[:,:,0] = b
        br[:,:,1] = g
        br[:,:,2] = 255 - r
    
        # Muestra composicion
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Transformando la imagen capturada en una colección de círculos, cuyo radio depende de la intensidad promedio de la zona

```python
vid = cv2.VideoCapture(0)

ncells = 10
off = int(ncells/2)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()
    

    if ret:
        #Dimensiones originales
        h, w, c = frame.shape
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w/ncells),int(h/ncells)),cv2.INTER_NEAREST)
        #Dimensiones reducidas
        h2, w2, c2 = down_frame.shape

        #Separamos canales
        r = down_frame[:,:,0]
        g = down_frame[:,:,1]
        b = down_frame[:,:,2]

        # Creamos imagen negra
        gris_up_frame = np.zeros((h2*ncells,w2*ncells,1), dtype = np.uint8)

        for y in range(0,h2):
            for x in range(0,w2):
                # La suma de los valores RGB define el radio del círculo
                rad = int ( (r[y,x] + g[y,x] + b[y,x])/(ncells*3*2))
                cv2.circle(gris_up_frame,(x*ncells+off, y*ncells + off), rad, (255,255,255), -1)

    
        # Muestra fotograma resultante
        cv2.imshow('Cam', gris_up_frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```
